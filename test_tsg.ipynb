```python
import tsaug
from tsaug.visualization import plot
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
random_signal = [ math.cos(i) for i in range(100)]
random_signal = np.array(random_signal)
X = range(100)
X = np.array(X)
plt.plot(X, random_signal)
plt.show()
```

```python
random_signal2,X2 = tsaug.AddNoise(loc=0,scale=1).augment(random_signal, X)
plt.plot(X2, random_signal2)
plt.show()
```

```python
random_signal2,X2 = tsaug.Crop(size=100).augment(random_signal, X)
plt.plot(X2, random_signal2)
plt.show()
```

```python
random_signal2,X2 = tsaug.Drift(max_drift=0.1, n_drift_points=10).augment(random_signal, X)
plt.plot(X2, random_signal2)
plt.show()
```

```python
random_signal2,X2 = tsaug.TimeWarp(n_speed_change=10, max_speed_ratio=10).augment(random_signal, X)
plt.plot(X2, random_signal2)
plt.show()
```
